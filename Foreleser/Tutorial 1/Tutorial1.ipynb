## a)

```python
#Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
#Creating data
x = np.array([1, 2, 3, 4, 5])
y = np.array([0.83, 6.99, 7.57, 5.99, 9.84])
n = x.size
```

```python
#Calculating a1
frac1 = n * (x * y).sum() - y.sum() * x.sum()
frac2 = n * (x**2).sum() - x.sum()**2
a1 = frac1 / frac2
```

```python
#Calculating a0
a0 = y.mean() - a1 * x.mean()
```

```python
#Function
f = a0 + a1 * x
print(f)
```

## b)

```python
df = pd.DataFrame({'x' : x, 'y' : y})
model1 = smf.ols('y ~ x', data = df)
m1 = model1.fit()
print(m1.params)

df_new = pd.DataFrame({'x' : [3], 'y' : [0]})
pred1 = m1.predict(df_new)
print(f'{pred1[0]:.3f}')
```

## c)

```python
#Prediction for diffrent K
x0 = 3
distances = np.abs(x - x0)
print(distances)
```

```python
#Sort distances by index
sorted_idx = np.argsort(distances)
print(sorted_idx)
```

```python
#K = 1
k1 = np.mean(y[sorted_idx[:1]])
print(k1)
```

```python
#K = 3
k3 = np.mean(y[sorted_idx[:3]])
print(round(k3, 2))
```

```python
#K = 5
k5 = np.mean(y[sorted_idx[:5]])
print(round(k5, 2))
```

```python
#Print results
print(f'Predicted value for K = 1 is {k1:.2f}')
print(f'Predicted value for K = 3 is {k3:.2f}')
print(f'Predicted value for K = 5 is {k5:.2f}')
```

## d)

```python
#Function for KNN
def knn(x0, x, y, K = 20):
    d = np.abs(x - x0)
    o = np.argsort(d)[:K]
    ypred = np.mean(y[o])
    return ypred
```

```python
#Experimenting with the function
K = 3
x0 = 3
d = np.abs(x - x0)
o = np.argsort(d)[:K]
ypred = np.mean(y[o])
print(round(ypred, 2))
```
