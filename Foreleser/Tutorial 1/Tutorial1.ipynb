```python
#Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
#Creating data
x = np.array([1, 2, 3, 4, 5])
y = np.array([0.83, 6.99, 7.57, 5.99, 9.84])
n = x.size
```

```python
#Calculating a1
frac1 = n * (x * y).sum() - y.sum() * x.sum()
frac2 = n * (x**2).sum() - x.sum()**2
a1 = frac1 / frac2
```

```python
#Calculating a0
a0 = y.mean() - a1 * x.mean()
```

```python
#Function
f = a0 + a1 * x
print(f)
```

```python
df = pd.DataFrame({'x' : x, 'y' : y})
model1 = smf.ols('y ~ x', data = df)
m1 = model1.fit()
print(m1.params)

df_new = pd.DataFrame({'x' : [3], 'y' : [0]})
pred1 = m1.predict(df_new)
print(f'{pred1[0]:.3f}')
```

## c)

```python
#Prediction for different K
x0 = 3
distances = np.abs(x - x0)
print(distances)
```

```python
#Sort distances by index
sorted_idx = np.argsort(distances)
print(sorted_idx)
```

```python
#K = 1
k1 = np.mean(y[sorted_idx[:1]])
print(k1)
```

```python
#K = 3
k3 = np.mean(y[sorted_idx[:3]])
print(round(k3, 2))
```

```python
#K = 5
k5 = np.mean(y[sorted_idx[:5]])
print(round(k5, 2))
```

```python
#Print results
print(f'Predicted value for K = 1 is {k1:.2f}')
print(f'Predicted value for K = 3 is {k3:.2f}')
print(f'Predicted value for K = 5 is {k5:.2f}')
```

## d)

```python
#Function for KNN
def knn(x0, x, y, K = 20):
    d = np.abs(x - x0)
    o = np.argsort(d)[:K]
    ypred = np.mean(y[o])
    return ypred
```

## a)

```python
#Loading data
from ISLP import load_data

College = load_data('College')
College.describe(include = 'all')
```

## b)
***

```python
#Dividing data 50/50 into trainind and test sets
np.random.seed(123)
n = College.shape[0]
train_idx = np.random.choice(n, size = n // 2, replace = False)
train = College.iloc[train_idx]
test = College.drop(College.index[train_idx])
```

## c)
***

```python
#Fitting linear regression model
model2 = smf.ols('Apps ~ Private + Accept', data = train)
m2 = model2.fit()
print(m2.summary())
```

## d)
***

```python
#Predicting on training data
pred_train = m2.predict(train)

#Compute MSE
mse_train = ((train['Apps'] - pred_train) ** 2).mean()
print(round(mse_train, 4))
```

## e)
***

```python
#C with only Accept as predictor
model3 = smf.ols('Apps ~ Accept', data = train)
m3 = model3.fit()
print(m3.summary())
```

```python
#MSE for M3
pred_train_m3 = m3.predict(train)
mse_train_m3 = ((train['Apps'] - pred_train_m3) ** 2).mean()
print(round(mse_train_m3, 4))
```

```python
#RMSE for M2 and M3
rmse_m2 = np.sqrt(mse_train)
rmse_m3 = np.sqrt(mse_train_m3)
print(f'RMSE for M2: {rmse_m2:.4f}')
print(f'RMSE for M3: {rmse_m3:.4f}')
```

We clearly see that MSE for both `Accept` and `Private` yields the lower MSE, giving a RMSE of 1 272 against M3's RMSE of 1 273.

## f)
***

```python
#Test-MSE for M2 and M3
pred_test_m2 = m2.predict(test)
mse_test_m2 = ((test['Apps'] - pred_test_m2) ** 2).mean()
print(f'Test MSE for M2: {mse_test_m2:.4f}')

pred_test_m3 = m3.predict(test)
mse_test_m3 = ((test['Apps'] - pred_test_m3) ** 2).mean()
print(f'Test MSE for M3: {mse_test_m3:.4f}')
```

As we can see, test MSE for M3 is lower than that of M2, suggesting that the predictor `Private` adds more noise than signal — the simpler model generalises better to new data. For both models, test-MSE is higher than training-MSE, as the models are optimised on training data and will always fit it better than unseen data.

## g)
***

```python
#Experimenting with different random seeds
for seed in [1, 42, 99, 200, 999]:
    np.random.seed(seed)
    n = College.shape[0]
    train_idx = np.random.choice(n, size = n // 2, replace = False)
    train = College.iloc[train_idx]
    test = College.drop(College.index[train_idx])

    m2 = smf.ols('Apps ~ Private + Accept', data = train).fit()
    m3 = smf.ols('Apps ~ Accept', data = train).fit()

    mse_test_m2 = ((test['Apps'] - m2.predict(test)) ** 2).mean()
    mse_test_m3 = ((test['Apps'] - m3.predict(test)) ** 2).mean()

    print(f'Seed {seed}: M2 test-MSE = {mse_test_m2:.0f},  M3 test-MSE = {mse_test_m3:.0f}')
```

## h)
***

```python
#KNN for Apps
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
```

```python
X_train = train[['Accept']].values
X_test = test[['Accept']].values
y_train = train['Apps'].values
y_test = test['Apps'].values
```

```python
#Standardising the predictor
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#KNN with multiple K values
for k in [1, 5, 10, 20, 50]:
    knn_model = KNeighborsRegressor(n_neighbors = k)
    knn_model.fit(X_train_scaled, y_train)
    pred_knn = knn_model.predict(X_test_scaled)
    mse_knn = ((y_test - pred_knn) ** 2).mean()
    print(f'KNN with K = {k}: Test MSE = {mse_knn:.0f}')
```

KNN performs worse than linear regression, as it misses heavily for universities with high application rates but similar acceptance rates. Linear regression follows the estimated line and handles these extreme values better. This is further highlighted by the $R^2$ of 0.878, indicating a clear linear relationship between Accept and Apps.
