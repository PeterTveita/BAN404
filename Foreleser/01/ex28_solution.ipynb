## f)
***

```python
#New variable Elite
college['Elite'] = pd.cut(college['Top10perc'], [0, 50, 100], labels = ['No', 'Yes'])
college
```

```python
#Amount of elite colleges
college['Elite'].value_counts()
```

```python
#Boxplot Outstate vs Elite
college.boxplot(column='Outstate', by='Elite')
plt.xlabel('Elite Status')
plt.ylabel('Out-of-State Tuition')
plt.title('Out-of-State Tuition by Elite Status')
plt.suptitle('')  # Remove the automatic title
plt.show()
```

## g)
***


```python
#Subplots for quantitav variables
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0, 0].hist(college['Apps'], bins = 20, color = 'lightblue', edgecolor = 'white')
ax[0, 0].set_title('Distribution of Applications')
ax[0, 0].set_xlabel('Number of Applications')
ax[0, 0].set_ylabel('Frequency')

ax[0, 1].hist(college['Accept'], bins = 20, color = 'lightgreen', edgecolor = 'white')
ax[0, 1].set_title('Distribution of Acceptances')
ax[0, 1].set_xlabel('Number of Acceptances')
ax[0, 1].set_ylabel('Frequency')

ax[1, 0].hist(college['Enroll'], bins = 15, color = 'lightcoral', edgecolor = 'white')
ax[1, 0].set_title('Distribution of Enrollments')
ax[1, 0].set_xlabel('Number of Enrollments')
ax[1, 0].set_ylabel('Frequency')

ax[1, 1].hist(college['Outstate'], bins = 20, color = '#B8F2E6', edgecolor = 'white')
ax[1, 1].set_title('Distribution of Out-of-State Tuition')
ax[1, 1].set_xlabel('Out-of-State Tuition')
ax[1, 1].set_ylabel('Frequency')

plt.show()
```
